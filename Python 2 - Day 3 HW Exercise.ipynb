## Exercise 1: 
Create an abstract class called Vehicle with 
- color, make, and year private attributes 

and abstract method called 
- get_info() which returns all information
- honk() which returns a string of honking sound

Create Car, Boat, and Bicycle class which is a child class of the Vehicle class.  Then, create an instance of each child class and call get_info() method to print out its information in a user friendly format. Add other methods as you like.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def get_info(self):
        pass
    def honk(self):
        pass
    
class Car(Vehicle):
    def __init__ (self, color, make, year):
        self.__color = color
        self.__make = make
        self.__year = year
        
    def get_info(self):
        return f'The car you own is {self.__color}, made by {self.__make}, produced in the year {self.__year}.'
        
    def honk(self):
        return 'BEEEEEP!'
        
class Boat(Vehicle):
    def __init__(self, color, make, year):
        self.__color = color
        self.__make = make
        self.__year = year
    def get_info(self):
        return f'The boat you own is {self.__color}, made by {self.__make}, its maiden voyage was in the year {self.__year}.'
        
    def honk(self):
        return 'TOOOTOOOT!'
        
class Bicycle(Vehicle):
    def __init__(self, color, make, year):
        self.__color = color
        self.__make = make
        self.__year = year

    def get_info(self):
        return f'The bicycle you own is {self.__color}, made by {self.__make}, produced in the year {self.__year}.'
        
    def honk(self):
        return 'DING! DING!'
        
c1 = Car('Yellow', 'Honda', 2018)
b1 = Boat('Ocean Blue','MasterCraft', 2001)
bi1 = Bicycle('Blood Red', 'Mongoose', 1998)

my_vehicles = [c1,b1,bi1]
for i in my_vehicles:
    print(i.get_info() + '\n\tThis vehicle says ' + i.honk())
    
#print(c1.get_info()+ '\n'+c1.honk())
#print(b1.get_info() + '\n'+ b1.honk())
#print(bi1.get_info()+ '\n'+ bi1.honk())
```

## Exercise 2: 
Create an abstract class called Person with name and address attributes, and print_info() abstract method. 
1. Build an Employee class inherited from Person class that 
    - stores hired date and salary attributes 
    - print_info() method that will be used to print out all information about an employee
    - a method to calculate employee's monthly pay where federal tax is 15%, MD tax is 4.5%, and a local tax is 3%.
    
2. Build a Customer class also inherited from Person class that
    - stores balance attribute
    - print_info() method that will be used to print out all information about a customer


Build any method you need but apply encapsulation at all possible. Then, write a program to show your classes work with multiple employees and customers. 

CHALLENGE!!!
Create an Address class and use it in Person class

```python
from abc import ABC, abstractmethod

class Person(ABC):
    @abstractmethod
    def print_info():
        pass
    

class Employee(Person):
    def __init__(self, name, address, sdate, salary):
        self.__name = name
        self.__address = address
        self.__sdate = sdate
        self.__salary = salary
    
    def salary_taxed(self):
        ftax = int(self.__salary)*.15
        print(f'\nFederal tax witheld {ftax}')
        mdtax = int(self.__salary)*.045
        print(f'Maryland tax witheld {mdtax}')
        ltax = int(self.__salary)*.03
        print(f'Local tax witheld {ltax}')
        return f'Salary after taxes {self.__salary - (ftax+mdtax+ltax)}\n'
        
    def print_info(self):
        print(f'Employee name {self.__name}.')
        print(f'Employee address {self.__address}.')
        print(f'Employee start date {self.__sdate}')
        print(f'Employee monthly salary before taxes {self.__salary}\n')
        

        
        
            
        
class Customer(Person):
    def __init__(self, name, address, balance):
        self.__name = name
        self.__address = address
        self.__balance = balance
    
    def print_info(self):
        print(f'Customer name {self.__name}.')
        print(f'Customer address {self.__address}.')
        print(f'Customer balance is {self.__balance}\n')

    
    
e1 = Employee('Fred','200 Endall circle', '9/20/2020', 100000)
e2 = Employee('Trinity', '1 Zion base', '01/01/2139', 4750)
print(e1.print_info())
print(e1.salary_taxed())

print(e2.print_info())
print(e2.salary_taxed())

c1 = Customer('Tammy', '1900 oldoaks rd.', 17550)
c2 = Customer('Styles', '987 something ct', 1578900)
print(c1.print_info())
print(c2.print_info())
```
