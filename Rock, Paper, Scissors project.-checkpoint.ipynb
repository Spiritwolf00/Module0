```python
import random as ran
#rock paper scissors:
player_score = 0
pc_score = 0
tied_score = 0
x='y'
while x != 'n':
    player_choice = ''
    print(player_choice)
    player_choice = input(f'Choose rock, paper, or scissors: ')
    player_choice = player_choice.lower()
    print(f'You chose {player_choice}!')
    pc_choice = ran.randint(1,3)
    if pc_choice == 1:
        pc_choice = 'rock'
    elif pc_choice == 2:
        pc_choice = 'paper'
    else:
        pc_choice = 'scissors'
    print(f'\tThe PC chose {pc_choice}!')
    if player_choice == 'rock' and pc_choice == 'scissors':
        print('rock beats scissors! You win!')
        player_score += 1
    elif player_choice == 'scissors' and pc_choice =='rock':
        print('rock beats scissors! The PC wins!')
        pc_score += 1
    
    elif player_choice == 'scissors' and pc_choice == 'paper':
        print('scissors beats paper! You win!')
        player_score += 1
    elif player_choice == 'paper' and pc_choice =='scissors':
        print('scissors beats paper! The PC wins!')
        pc_score += 1
            
    elif player_choice == 'paper' and pc_choice == 'rock':
        print('paper beats rock! You win!')
        player_score += 1
    elif player_choice == 'rock' and pc_choice =='paper':
        print('paper beats rock! The PC wins!')
        pc_score += 1
            
    elif player_choice == pc_choice:
        print('You and the PC tied!')
        tied_score += 1
        
    else:
        print('please only enter "rock" "paper" or "scissors" only.')
    x = input('would you like to play again? type "n" to quit, hit return to keep playing: ')

print(f'I hope you had fun, you won {player_score} rounds, and the PC won {pc_score} rounds.')
print(f'You and the PC tied {tied_score} times.')
print('Thanks for playing.')
```
